```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://quotes.toscrape.com/"
page = requests.get(URL)
#print("\n***Page***")
#print(page.text)
#print("***END***")
```

```python
soup = BeautifulSoup(page.content, "html.parser")
#print ("\n***Soup***")
#print(soup)
#print("***END***")
```

```python
results = soup.find(id="ResultsContainer")
#print ("\n***results***")
#print(results.prettify())
#print("***END***")
```

```python
print ("\n***QUOTES OF THE DAY***")
quote_elements = soup.find_all("div", class_="quote")
for quote_elements in quote_elements:
    text_element = quote_elements.find("span", class_="text")
    author_element = quote_elements.find("small", class_="author")
    print(text_element)
    print(author_element)
    print()
    print(text_element.text)
    print(author_element.text)
    print()
    print(text_element.text.strip())
    print(author_element.text.strip())
    print()
print("***END***")
```

```python
quote_elements = soup.find_all("div", class_="quote")

count = 0

for quote in quote_elements:
    tags = quote.find_all("a", class_="tag")
    tag_texts = [tag.text.lower() for tag in tags]

    if "inspirational" in tag_texts:
        count += 1

        text_element = quote.find("span", class_="text")
        author_element = quote.find("small", class_="author")

        if text_element and author_element:
            print(text_element.text.strip())
            print(author_element.text.strip())
            print()

print("Number of inspirational quotes:", count)
print("***END***")
```

```python
from collections import defaultdict

url = "https://quotes.toscrape.com/page/1/"

author_count = defaultdict(int)

while url:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quote_elements = soup.find_all("div", class_="quote")

    for quote in quote_elements:
        author = quote.find("small", class_="author").get_text(strip=True)
        author_count[author] += 1

    # go to next page
    next_btn = soup.find("li", class_="next")
    if next_btn:
        next_url = next_btn.find("a")["href"]
        url = "https://quotes.toscrape.com" + next_url
    else:
        url = None

# print results
for author, count in author_count.items():
    print(author, ":", count)

print("***END***")
```
